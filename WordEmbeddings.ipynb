```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import pickle  
```

```python
word_embeddings = pickle.load( open( "word_embeddings_subset.p", "rb" ) )
print(len(word_embeddings))
print(type(word_embeddings))
```

```python
countryVector = word_embeddings['country'] 
print(type(countryVector)) 
print(countryVector)  
```

```python
def vec(w):
    return word_embeddings[w]
```

```python
words = ['oil', 'gas', 'happy', 'sad', 'city', 'town', 'village', 'country', 'continent', 'petroleum', 'joyful']

bag2d = np.array([vec(word) for word in words]) # Convert each word to its vector representation
print(bag2d.shape) # 11 rows for different words and 300 columns for each of its vector representation 

plt.figure(figsize=(10,10))

# Choosing any 2 columns
col1 = 3 
col2 = 2 

# Print an arrow for each word
for word in bag2d:
    plt.arrow(0, 0, word[col1], word[col2], head_width=0.005, head_length=0.005, ec='r', width = 1e-5)

    
plt.scatter(bag2d[:, col1], bag2d[:, col2]); # Plot a dot for each word

# Add the word label over each dot in the scatter plot
for i in range(0, len(words)):
    plt.annotate(words[i], (bag2d[i, col1], bag2d[i, col2]))

plt.show()
```

```python
words = ['sad', 'happy', 'town', 'village']

bag2d = np.array([vec(word) for word in words]) # Convert each word to its vector representation

plt.figure(figsize=(10,10))

col1 = 3
col2 = 2

# Print an arrow for each word
for word in bag2d:
    plt.arrow(0, 0, word[col1], word[col2], head_width=0.0005, head_length=0.0005, fc='r', ec='r', width = 1e-5)
    
# print the vector difference between village and town
village = vec('village')
town = vec('town')
diff = town - village
plt.arrow(village[col1], village[col2], diff[col1], diff[col2], fc='b', ec='b', width = 1e-5)

# print the vector difference between village and town
sad = vec('sad')
happy = vec('happy')
diff = happy - sad
plt.arrow(sad[col1], sad[col2], diff[col1], diff[col2], fc='b', ec='b', width = 1e-5)


plt.scatter(bag2d[:, col1], bag2d[:, col2]); # Plot a dot for each word

# Add the word label over each dot in the scatter plot
for i in range(0, len(words)):
    plt.annotate(words[i], (bag2d[i, col1], bag2d[i, col2]))


plt.show()
```
